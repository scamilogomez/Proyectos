```
import ipywidgets as widgets
from IPython.display import display

# Funciones que se ejecutan al presionar los botones
def imprimir_letras1(boton):
    letras = "ABC"
    print(letras)

def imprimir_letras2(boton):
    letras = "XYZ"
    print(letras)

# Crear los botones y asociar las funciones
boton1 = widgets.Button(description="Botón 1")
boton1.on_click(imprimir_letras1)

boton2 = widgets.Button(description="Botón 2")
boton2.on_click(imprimir_letras2)

# Mostrar los botones
display(widgets.VBox([boton1, boton2]))
```
