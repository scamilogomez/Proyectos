```python
import ipywidgets as widgets

class NumerosAlternados:
    def __init__(self):
        # Crear botones
        self.boton1 = widgets.Button(description="Boton 1")
        self.boton2 = widgets.Button(description="Boton 2")

        # Crear etiqueta
        self.etiqueta = widgets.Label(value="Presione un boton")

        # Agregar eventos a los botones
        self.boton1.on_click(self.mostrar_numero1)
        self.boton2.on_click(self.mostrar_numero2)

        # Crear panel y agregar componentes
        self.panel = widgets.VBox([self.boton1, self.boton2, self.etiqueta])

        # Inicializar contadores
        self.contador1 = 0
        self.contador2 = 0

    def mostrar_numero1(self, boton):
        self.etiqueta.value = "Boton 1: " + str(self.contador1)
        self.contador1 += 1

    def mostrar_numero2(self, boton):
        self.etiqueta.value = "Boton 2: " + str(self.contador2)
        self.contador2 += 1

# Crear objeto NumerosAlternados
app = NumerosAlternados()

# Mostrar panel
display(app.panel)

```
